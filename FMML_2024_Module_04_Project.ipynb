Creating a function for the derivative of $f(x)$ called df(x)

````python
def df(x):
    return 2 * x + 1
````

### Gradient Descent

````python
##########################
##########################

## TASK-1 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

##########################
##########################

# Gradient Descent
# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(new_x)  # Calculate the gradient (derivative) at new_x
    new_x = previous_x - step_multiplier * gradient  # Update new_x using the gradient and step size

    step_size = abs(new_x - previous_x)
    # print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))  # Append the gradient at the new_x

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
````

## Gradient Descent

````python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
````

````python
## TASK-8 : Code the same things as they were done for the first two examples
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

# 1. Define the function h(x, y)
def h(x, y):
    return x**2 + y**2 + 2*x  # Example function, you can change it

# 2. Define partial derivatives
def dh_dx(x, y):
    return 2*x + 2

def dh_dy(x, y):
    return 2*y

# 3. Make data for plotting
x_4 = np.linspace(-5, 5, 100)
y_4 = np.linspace(-5, 5, 100)
x_4, y_4 = np.meshgrid(x_4, y_4)

# 4. Plot the function and its derivatives (3D plot)
fig = plt.figure(figsize=(12, 6))

# Subplot 1: 3D plot of h(x, y)
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(x_4, y_4, h(x_4, y_4), cmap=cm.viridis)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('h(x, y)')
ax1.set_title('Cost Function')

# Subplot 2: Contour plot of h(x, y) (optional)
ax2 = fig.add_subplot(122)
contour = ax2.contour(x_4, y_4, h(x_4, y_4), cmap=cm.viridis)
ax2.clabel(contour, inline=1, fontsize=8)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Contour Plot')

plt.tight_layout()
plt.show()

# 5 & 6. Gradient Descent and Visualization (adapt for 2 variables)
# You'll need to adjust the gradient_descent function for 2 variables
# and the visualization to plot on the 3D surface or contour plot
# I'll provide a basic adaptation of gradient_descent:

def gradient_descent_2d(
    derivative_func_x, derivative_func_y, initial_guess, multiplier=0.02, precision=0.001, max_iter=300
):
    new_x, new_y = initial_guess
    x_list, y_list = [new_x], [new_y]

    for n in range(max_iter):
        prev_x, prev_y = new_x, new_y
        gradient_x = derivative_func_x(prev_x, prev_y)
        gradient_y = derivative_func_y(prev_x, prev_y)
        new_x = prev_x - multiplier * gradient_x
        new_y = prev_y - multiplier * gradient_y

        step_size = abs(new_x - prev_x) + abs(new_y - prev_y)  # Combined step size
        x_list.append(new_x)
        y_list.append(new_y)

        if step_size < precision:
            break

    return new_x, new_y, x_list, y_list

# Example usage:
initial_guess = (2, 2)  # Starting point
local_min_x, local_min_y, x_list, y_list = gradient_descent_2d(dh_dx, dh_dy, initial_guess)

print("Local minimum occurs at x:", local_min_x, "y:", local_min_y)

# Plot the gradient descent path on the contour plot
ax2.plot(x_list, y_list, color='red', marker='o', markersize=5, label='Gradient Descent Path')
ax2.legend()
plt.show()
````

#### Mean Squared Error: a cost function for regression problems

#### $$RSS = \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
#### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
#### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y - \hat{y} \big)^2 $$

````python
# Make sample data
x_5 = np.array([[0.1, 1.2, 2.4, 3.2, 4.1, 5.7, 6.5]]).transpose()
y_5 = np.array([1.7, 2.4, 3.5, 3.0, 6.1, 9.4, 8.2]).reshape(7, 1)

print('Shape of x_5 array:', x_5.shape)
print('Shape of y_5 array:', y_5.shape)
````

````python
# Challenge: Write a python function mse(y, y_hat) returns the MSE? Call the mse(y, y_hat)
# function and print out the MSE for the y_hat calculated above.

def mse(y, y_hat):
    #mse_calc = 1/7 * sum((y - y_hat)**2)
    #mse_calc = (1/y.size) * sum((y - y_hat)**2)
    mse_calc = np.average((y - y_hat)**2, axis=0)
    return mse_calc
````

````python
# Plotting MSE
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

# Example: Define the MSE cost function (for a simple linear regression)
def mse(theta0, theta1, X, y):
    predictions = theta0 + theta1 * X
    return np.mean((predictions - y) ** 2)

# Example data for X and y (replace with your data)
X = np.linspace(-5, 5, 100)
y = 2 * X + 1 + np.random.normal(0, 2, 100)  # Linear relation with noise

# Generate theta0 and theta1 values (plot_t0 and plot_t1)
theta0_vals = np.linspace(-10, 10, 100)
theta1_vals = np.linspace(-10, 10, 100)

# Create meshgrid for theta0 and theta1
plot_t0, plot_t1 = np.meshgrid(theta0_vals, theta1_vals)

# Compute the cost (MSE) for each (theta0, theta1) combination
plot_cost = np.array([[mse(t0, t1, X, y) for t0 in theta0_vals] for t1 in theta1_vals])

# Plotting MSE
fig = plt.figure(figsize=[16, 12])
ax = fig.add_subplot(111, projection='3d')  # Correct way to create a 3D axis

ax.set_xlabel('Theta 0', fontsize=20)
ax.set_ylabel('Theta 1', fontsize=20)
ax.set_zlabel('Cost - MSE', fontsize=20)

ax.plot_surface(plot_t0, plot_t1, plot_cost, cmap=cm.hot)  # MSE surface plot with 'hot' colormap
plt.show()
````

#### Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

#### $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

#### $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

#### MSE & Gradient Descent

````python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
````

````python
# Assuming mse function is defined as:
def mse(X, y, theta0, theta1):
    predictions = theta0 + theta1 * X
    return np.mean((predictions - y) ** 2)

# Initialize parameters
multiplier = 0.01
thetas = np.array([2.9, 2.9])

# Collect data points for scatter plot
plot_vals = thetas.reshape(1, 2)
mse_vals = mse(x_5, y_5, thetas[0], thetas[1])

# Gradient Descent Loop
for i in range(1000):
    # Gradient descent step: Update thetas
    thetas = thetas - multiplier * grad(x_5, y_5, thetas)

    # Append the new values to our numpy arrays
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
    mse_vals = np.append(arr=mse_vals, values=mse(x_5, y_5, thetas[0], thetas[1]))

# Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('MSE is:', mse(x_5, y_5, thetas[0], thetas[1]))
````
